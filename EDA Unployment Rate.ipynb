**EDA Unployment Rate**

```python
import numpy                 as np
import pandas                as pd
import matplotlib.pyplot     as plt
import seaborn               as sns
```

```python
df = pd.read_csv('Cifras_Desempleo.csv',encoding = 'UTF-8',sep=";", decimal = ",")
df.head(30)
```

```python
# Cambio de numeración de los meses del año
df = df.rename(columns={'1':'01','2':'02','3':'03','4':'04','5':'05','6':'06','7':'07','8':'08','9':'09'})
```

```python
df.info()
```

```python
df.shape
```

**TD** : Es la relación porcentual entre el número de personas que están buscando trabajo –DS-:, y el número de personas que integran la fuerza laboral

```python
# Filtrar por los conceptos básicos que sustentan las cifra de empleo
df_filter = df[(df["Concepto"] == "TD")]
df_filter
```

```python
df_filter.shape
```

```python
# Mejorar el formato de tabla
Months = ["01",'02','03','04','05','06','07','08','09','10','11','12']
df_filter = pd.melt(df_filter, id_vars = ["Ciudad" , "Concepto", "Año"], value_vars = Months ,var_name = "Month", value_name = "Valor" )
df_filter
```

```python
# Reemplazar en la columna valor comas por punto
df_filter["Valor"] = [x.replace(',', '.') for x in df_filter["Valor"]]
```

```python
# Formatear las columnas Año, Month y Valor
df_filter["Año"]= df_filter["Año"].astype(str)
df_filter["Month"]= df_filter["Month"].astype(str)
df_filter["Valor"]= df_filter["Valor"].astype(float)
```

```python
# Filtar por cada uno de los conceptos básicos y así sacar las estadísticas básicas 
df_filter_TD = df_filter[(df_filter["Concepto"] == "TD")]
```

**Unemployment Rate Statistics**

```python
df_filter_TD
```

```python
df_filter_TD.describe()
```

```python
# Crear una columna con la unión de año y mes
df_filter_TD['Year_Month'] = df_filter_TD['Año']+"-"+df_filter_TD['Month']
df_filter_TD['Year_Month']
```

```python
TD_by_year = df_filter_TD.groupby(["Year_Month"]).mean().reset_index()
TD_by_year["Ciudad"] = "Nacional"
TD_by_year
```

```python
# Tasa de desempleo por año
plt.plot (TD_by_year["Year_Month"],TD_by_year['Valor'],linewidth = '5')
plt.title("Unployment Rate by Year")
plt.rcParams["figure.figsize"] = (20, 20)
plt.xticks(rotation = 90)
plt.grid(True)
```

```python
TD_by_year_and_City= df_filter_TD.groupby(["Year_Month" ,"Ciudad"]).mean().reset_index()
TD_by_year_and_City
```

```python
g = sns.lineplot(data=df_filter, x = "Año", y= "Valor", hue = "Ciudad", palette ="Set2", sizes=(.25, 100))
#plt.gcf().set_size_inches(15, 8)
#plt.xticks(rotation=45)
```

```python
# Mapa de calor de la TD por año y ciudad
TD_wide = TD_by_year_and_City.pivot("Año","Ciudad","Valor")
sns.heatmap (TD_wide, linewidths =0.1,cmap = 'YlGnBu',fmt="d",cbar_kws={"orientation": "vertical"})
```

**Ciudades con tasas de desempleo más altas**

```python
# Filtro por cada una de las 5 ciudades con más desempleo en los últimos años
df_Armenia = df_filter_TD[(df_filter_TD["Ciudad"] == "Armenia")]
df_Quibdo = df_filter_TD[(df_filter_TD["Ciudad"] == "Quibdó")]
df_Cucuta = df_filter_TD[(df_filter_TD["Ciudad"] == "Cúcuta A.M.")]
df_Riohacha = df_filter_TD[(df_filter_TD["Ciudad"] == "Riohacha")]
df_Valledupar = df_filter_TD[(df_filter_TD["Ciudad"] == "Valledupar")]
```

```python
# Comparación de las ciudades con el promedio nacional
Merge= pd.concat([df_Armenia,df_Quibdo,df_Cucuta,df_Riohacha,df_Valledupar,TD_by_year]).reset_index()
g = sns.FacetGrid(Merge,col="Ciudad", height = 4, col_wrap=1, aspect=2)
g.map_dataframe(sns.barplot,x ="Year_Month", y= "Valor")
```
